```python
#Plotting function
def plot_feature_importances(features, title):
    sorted_features = dict(sorted(features.items(), key=lambda item: item[1], reverse=True))
    plt.figure(figsize=(10, 8))
    plt.bar(list(sorted_features.keys()), list(sorted_features.values()))
    plt.ylabel('Feature Importance')
    plt.title(title)
    plt.xticks(rotation=90)
    plt.show()

# Plotting the charts
plot_feature_importances(features_top10, 'Top 10 Feature Importances')
plot_feature_importances(features_top20, 'Top 20 Feature Importances')
plot_feature_importances(features_top30, 'Top 30 Feature Importances')
plot_feature_importances(features_top41, 'Top 41 Feature Importances')
```
